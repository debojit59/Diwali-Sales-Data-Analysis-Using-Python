```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv("Diwali Sales Data.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Droping BlaNk Columns Status and Unnamed1
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#dropping Null Values from the data set
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head(10)
```

```python
#chaning the data type of column called "Amount" from Float to Int
```

```python
df['Amount']=df['Amount'].astype("int")
```

```python
df['Amount'].dtype
```

```python
#chaning the name of column
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

**Gender**

```python
a=sns.countplot(x="Gender",data=df)
for bar in a.containers:
  a.bar_label(bar)
sns.set(rc={'figure.figsize':(7,5)})
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sales_gen
```

```python
b=sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bar in b.containers:
  b.bar_label(bar)
sns.set(rc={'figure.figsize':(7,5)})
```

From the avobe graphs we can understand that most of the buyers are femals not only that it is also observered that females are having more purchasing power than males


**AGE**

```python
df.columns
```

```python
axis=sns.countplot(data=df,x='Age Group',hue='Gender')
for bar in axis.containers:
  axis.bar_label(bar)
sns.set(rc={'figure.figsize':(7,5)})
```

```python
#Total amount vs Age Group
Sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
Sales_age
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Age Group',y='Amount', data=Sales_age)
```

From avobe graph we can realize that most of the buyers are from age group of 26-35


**STATE**

```python
df.columns
```

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

From the avobe graph we can see that unexpectedly most of the orders are from Uttarpradesh,Maharashtra and Karnataka respectively but total amount/sales is from UP,Karnataka,then Maharastra

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

From the above graph we can see that most of the buyers are married(women) and they have high purchasing power
